##### 3. What Assumptions Are Made by Linear Regression?

###### Answer

Linear Regression is a supervised machine learning algorithm used to predict continuous numerical values. To produce accurate and reliable predictions, Linear Regression relies on several assumptions. If these assumptions are violated, the model's performance may decrease, leading to inaccurate predictions and misleading conclusions.

The five main assumptions of Linear Regression are explained below.

---

##### 1. Linearity

The first assumption is that there should be a **linear relationship** between the independent variable(s) and the dependent variable. This means that any change in the independent variable should result in a proportional change in the dependent variable.

##### Example

If the number of hours studied increases, the exam marks are expected to increase in a straight-line pattern.

##### Diagram

##### Incorrect Pattern (Heteroscedasticity)

```text
Residuals

●
 ●
  ●
    ●
      ●
         ●      ●
-------------------------------> Predicted Values
```

---

##### 4. Normal Distribution of Residuals

The residuals (prediction errors) should follow a normal distribution with a mean close to zero. This assumption helps in making valid statistical inferences.

5. What do the slope and intercept represent in a linear regression model?

ANS  In a Linear Regression model, the relationship between the independent variable (**X**) and the dependent variable (**Y**) is represented by a straight-line equation. The two most important components of this equation are the **Slope** and the **Intercept**.

The equation of a Simple Linear Regression model is:

Formula

**Y = β₀ + β₁X**

Where:

- **Y** = Dependent Variable (Target)
- **X** = Independent Variable (Predictor)
- **β₀** = Intercept
- **β₁** = Slope (Regression Coefficient)

---

##### 1. Slope (β₁)

The **Slope** indicates how much the dependent variable (**Y**) changes when the independent variable (**X**) increases by one unit.

In simple terms, it tells us the **rate of change** between the two variables.

##### Example

Suppose the regression equation is:

**Marks = 20 + 5 × (Study Hours)**

Here,

- Intercept = 20
- Slope = 5

This means that **for every additional hour studied, the student's marks are expected to increase by 5 marks.**

##### Interpretation

- Positive Slope → Y increases as X increases.
- Negative Slope → Y decreases as X increases.
- Zero Slope → No relationship between X and Y.

---

#### 2. Intercept (β₀)

The **Intercept** is the predicted value of the dependent variable (**Y**) when the independent variable (**X**) is equal to zero.
It represents the point where the regression line crosses the **Y-axis**.

##### Example

Regression Equation:
**Marks = 20 + 5 × (Study Hours)**
When:
**Study Hours = 0**
Then,
**Marks = 20**
This means that if a student studies for **0 hours**, the predicted marks are **20**.

##### 6. What is the Purpose of the Cost Function in Regression?

#### Answer

The **cost function** is a mathematical function used to measure how well a regression model fits the training data. It calculates the difference between the **actual values** and the **predicted values** generated by the model.

The main purpose of the cost function is to **minimize prediction errors** by finding the best values of the model parameters (slope and intercept). During training, the algorithm continuously updates these parameters to reduce the cost function value, resulting in a more accurate regression model.

###### Formula (Mean Squared Error)

\[
J(\theta)=\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y_i})^2
\]

Where:

- **J(θ)** = Cost Function
- **n** = Number of observations
- **y** = Actual value
- **ŷ** = Predicted value

##### Working of Cost Function

```text
Training Data
      │
      ▼
Regression Model
      │
      ▼
Predicted Values
      │
      ▼
Calculate Error
(Actual - Predicted)
      │
      ▼
Cost Function
      │
      ▼
Update Model Parameters
```

###### Example

Suppose the actual house price is **₹50 lakh**, but the regression model predicts **₹47 lakh**. The cost function measures this error. The model then adjusts its parameters to reduce the difference and improve prediction accuracy.

##### 7. What is Mean Squared Error (MSE), and Why is it Commonly Used for Regression?

##### Answer

**Mean Squared Error (MSE)** is one of the most commonly used evaluation metrics in regression. It measures the average of the squared differences between the **actual values** and the **predicted values**. A lower MSE indicates that the regression model makes more accurate predictions.

---

##### Formula

\[
MSE=\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y_i})^2
\]

Where:

- **n** = Total number of observations
- **y** = Actual value
- **ŷ** = Predicted value

---

##### Working of MSE

```text
Actual Values
      │
      ▼
Predicted Values
      │
      ▼
Find Difference
(Actual − Predicted)
      │
      ▼
Square the Errors
      │
      ▼
Calculate Average
      │
      ▼
Mean Squared Error (MSE)
```

---

##### Example

Suppose the actual house prices are:

**Actual:** 50, 60, 70

**Predicted:** 48, 63, 68

Errors:

- (50 − 48)² = 4
- (60 − 63)² = 9
- (70 − 68)² = 4

\[
MSE=\frac{4+9+4}{3}=5.67
\]

Therefore, the **Mean Squared Error = 5.67**.

---

##### Why is MSE Commonly Used?

- It measures the average prediction error.
- Squaring the errors gives more importance to large errors.
- It is easy to calculate and interpret.
- It is differentiable, making it suitable for optimization algorithms like **Gradient Descent**.
- It helps compare the performance of different regression models.

**8. What is the Difference Between Mean Absolute Error (MAE), Mean Squared Error (MSE), and Root Mean Squared Error (RMSE)?**

**ANS:** Mean Absolute Error (MAE), Mean Squared Error (MSE), and Root Mean Squared Error (RMSE) are three commonly used evaluation metrics in regression. They measure the difference between the actual values and the predicted values, but each calculates the error differently.

### **Formula**

**Mean Absolute Error (MAE):**

\[
MAE=\frac{1}{n}\sum_{i=1}^{n}|y_i-\hat{y_i}|
\]

**Mean Squared Error (MSE):**

\[
MSE=\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y_i})^2
\]

**Root Mean Squared Error (RMSE):**

\[
RMSE=\sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y_i})^2}
\]

**Where:**

- **n** = Number of observations
- **y** = Actual value
- **ŷ** = Predicted value

### **Difference Between MAE, MSE and RMSE**

| **Basis** | **MAE** | **MSE** | **RMSE** |
|------------|---------|---------|----------|
| **Full Form** | Mean Absolute Error | Mean Squared Error | Root Mean Squared Error |
| **Calculation** | Average of absolute errors | Average of squared errors | Square root of MSE |
| **Formula** | \|Actual − Predicted\| | (Actual − Predicted)² | √MSE |
| **Effect of Large Errors** | Less sensitive | Highly sensitive | Moderately sensitive |
| **Unit** | Same as original data | Squared unit | Same as original data |
| **Interpretation** | Easy to understand | Gives more penalty to large errors | Easier to interpret than MSE |
| **Best Used When** | All errors are equally important | Large errors should be penalized | Measuring overall model performance |



9.What is the R-squared score, and how should it be interpreted?

**ANS:** The **R-squared (R²) score**, also known as the **Coefficient of Determination**, is a statistical measure used to evaluate the performance of a regression model. It indicates how well the independent variable(s) explain the variation in the dependent variable. The R² value ranges from **0 to 1** (or **0% to 100%**).

### **Formula**

\[
R^2 = 1 - \frac{\sum (y_i-\hat{y_i})^2}{\sum (y_i-\bar{y})^2}
\]

**Where:**

- **\(y_i\)** = Actual value
- **\(\hat{y_i}\)** = Predicted value
- **\(\bar{y}\)** = Mean of actual values

### **Interpretation of R² Score**

| **R² Score** | **Interpretation** |
|--------------|--------------------|
| **0** | The model explains none of the variation in the data. |
| **0.25** | Poor fit (25% of the variation is explained). |
| **0.50** | Moderate fit (50% of the variation is explained). |
| **0.75** | Good fit (75% of the variation is explained). |
| **1.00** | Perfect fit (100% of the variation is explained). |

### **Example**

Suppose a regression model predicts house prices and gives an **R² score of 0.85**.

This means that **85% of the variation in house prices is explained by the model**, while the remaining **15% is due to other factors or random errors**.

##### **Causes of Overfitting**

- Too many independent variables.
- Small training dataset.
- Presence of noise or outliers.
- Excessively complex regression model.

#### **How to Prevent Overfitting**

- Use more training data.
- Remove unnecessary features.
- Apply Regularization (Ridge or Lasso Regression).
- Perform Cross-Validation.
- Keep the model as simple as possible.

**13. What is Multicollinearity, and How Does It Affect Regression?**

**ANS:** **Multicollinearity** is a situation in which two or more independent variables in a regression model are highly correlated with each other. This means they provide similar information, making it difficult for the model to determine the individual effect of each variable on the dependent variable.

### **Example**

Suppose we are predicting house prices using:

- House Area
- Number of Rooms
- House Size (sq. ft.)

Here, **House Area** and **House Size** are highly related, causing multicollinearity.

### **Diagram**

```text
Area --------\
              \
               ---> House Price
              /
House Size --/
```

### **Effects of Multicollinearity**

- Makes regression coefficients unstable.
- Reduces model interpretability.
- Increases prediction errors.
- Makes feature importance difficult to determine.
- May reduce overall model reliability.

**17. How Does Lasso Regression Perform Feature Selection?**

**ANS:** **Lasso Regression (Least Absolute Shrinkage and Selection Operator)** is a regularization technique that performs **feature selection** by shrinking the coefficients of less important features to **zero**. Features with zero coefficients are automatically removed from the model, making it simpler and more efficient.

### **How Lasso Performs Feature Selection**

- Adds an **L1 penalty** to the cost function.
- Reduces the values of regression coefficients.
- Sets unimportant feature coefficients to **zero**.
- Removes unnecessary features automatically.
- Improves model performance and reduces overfitting.

### **Diagram**

```text
All Features
      │
      ▼
Lasso Regression
      │
      ▼
Important Features Kept
Unimportant Features Removed
```

### **Advantages**

- Performs automatic feature selection.
- Reduces overfitting.
- Creates simpler and more interpretable models.
- Improves prediction accuracy.

### **Conclusion**

Lasso Regression selects important features by shrinking the coefficients of less important variables to zero. This helps build a simpler, faster, and more accurate regression model.
