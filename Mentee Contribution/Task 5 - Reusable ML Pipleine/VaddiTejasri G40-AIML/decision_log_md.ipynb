
# Decision Log

Reusable Customer Churn Prediction Pipeline using scikit-learn

| Decision Area | Decision Taken | Reason |
|--------------|----------------|--------|
| Dataset | IBM Telco Customer Churn Dataset from Kaggle | A real-world customer churn dataset suitable for classification problems. |
| Removed Column | CustomerID | It is only a unique identifier and does not contribute to predicting customer churn. |
| Target Variable | Churn | This is the output variable indicating whether a customer leaves the service. |
| Features | Age, Gender, Tenure, MonthlyCharges, TotalCharges, ContractType, PaymentMethod, InternetService, SupportTickets | These customer-related features are useful for learning churn behavior. |
| Train-Test Split | 80:20 (test_size=0.2) | Ensures that the model is evaluated on unseen data while retaining enough data for training. |
| Random State | 42 | Produces the same train-test split every time for reproducibility. |
| Stratification | Used stratify=y | Maintains the same proportion of churn and non-churn customers in both training and testing datasets. |
| Missing Numerical Values | Median Imputation | Median is less sensitive to outliers than the mean. |
| Missing Categorical Values | Most Frequent Imputation | Preserves the most common category and prevents data loss. |
| Numerical Scaling | StandardScaler | Standardizes numerical features to improve model performance. |
| Categorical Encoding | OneHotEncoder(handle_unknown="ignore") | Converts categorical variables into numerical format while safely handling unseen categories. |
| Preprocessing | ColumnTransformer | Applies different preprocessing techniques to numerical and categorical columns efficiently. |
| Machine Learning Model | RandomForestClassifier | Performs well for classification problems, handles nonlinear relationships, and reduces overfitting using multiple decision trees. |
| Pipeline | sklearn Pipeline | Combines preprocessing and model training into one reusable workflow and prevents data leakage. |
| Evaluation Metrics | Accuracy, Precision, Recall, F1-score, Confusion Matrix | These metrics provide a complete understanding of classification performance. Recall is particularly important for identifying customers likely to churn. |
| Model Saving | joblib.dump() | Saves the complete preprocessing and trained model for future predictions without retraining. |
| Reusability | Loaded using joblib.load() | Enables prediction on new customer data using the same preprocessing and trained model. |



This project was developed using a reusable Scikit-learn Pipeline. All preprocessing operations, including missing value handling, feature scaling, categorical encoding, and model training, were integrated into a single pipeline. The pipeline was trained only on the training dataset to prevent data leakage and was saved using Joblib for future reuse.
