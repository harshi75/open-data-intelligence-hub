```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
cluster_features = train_data[[
    'Age',
    'Tenure',
    'Usage Frequency',
    'Support Calls',
    'Payment Delay',
    'Total Spend',
    'Last Interaction'
]]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

cluster_scaled = scaler.fit_transform(cluster_features)
```

```python
inertia = []

for k in range(2,11):

    model = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    model.fit(cluster_scaled)

    inertia.append(model.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(
    range(2,11),
    inertia,
    marker='o'
)

plt.title("Elbow Method")

plt.xlabel("Number of Clusters")

plt.ylabel("Inertia")

plt.show()
```

```python
kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

clusters = kmeans.fit_predict(cluster_scaled)
```

```python
train_data["Cluster"] = clusters
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    silhouette_score
)
```

```python
ridge_prediction = best_ridge.predict(X_test_scaled)

ridge_mae = mean_absolute_error(
    y_test,
    ridge_prediction
)

ridge_rmse = np.sqrt(
    mean_squared_error(
        y_test,
        ridge_prediction
    )
)

ridge_r2 = r2_score(
    y_test,
    ridge_prediction
)

print("Ridge Regression Results")
print("------------------------")
print("MAE:", round(ridge_mae, 4))
print("RMSE:", round(ridge_rmse, 4))
print("R² Score:", round(ridge_r2, 4))
```

```python
logistic_prediction = best_logistic.predict(
    X_test_scaled
)

logistic_accuracy = accuracy_score(
    y_test,
    logistic_prediction
)

logistic_precision = precision_score(
    y_test,
    logistic_prediction,
    zero_division=0
)

logistic_recall = recall_score(
    y_test,
    logistic_prediction,
    zero_division=0
)

logistic_f1 = f1_score(
    y_test,
    logistic_prediction,
    zero_division=0
)

print("Logistic Regression Results")
print("---------------------------")
print("Accuracy:", round(logistic_accuracy, 4))
print("Precision:", round(logistic_precision, 4))
print("Recall:", round(logistic_recall, 4))
print("F1 Score:", round(logistic_f1, 4))
```

```python
from sklearn.metrics import silhouette_score
import numpy as np

# Use a sample for faster silhouette calculation
sample_size = 10000

if len(cluster_scaled) > sample_size:
    random_indices = np.random.RandomState(42).choice(
        len(cluster_scaled),
        size=sample_size,
        replace=False
    )

    cluster_sample = cluster_scaled[random_indices]
    cluster_labels_sample = clusters[random_indices]

else:
    cluster_sample = cluster_scaled
    cluster_labels_sample = clusters

kmeans_silhouette = silhouette_score(
    cluster_sample,
    cluster_labels_sample
)

kmeans_inertia = kmeans.inertia_

print("K-Means Clustering Results")
print("--------------------------")
print("Number of Clusters:", kmeans.n_clusters)
print("Inertia:", round(kmeans_inertia, 2))
print("Silhouette Score:", round(kmeans_silhouette, 4))
```

#Report Conclusion

This project compared Ridge Regression, Logistic Regression, and
K-Means Clustering for customer churn analysis. Ridge Regression
was included to demonstrate a regression technique, although it is
not the most suitable approach for a binary target variable.
Logistic Regression was found to be the most appropriate predictive
model because customer churn contains two classes: churn and
non-churn. K-Means Clustering provided additional business value by
segmenting customers according to their behaviour. Therefore,
Logistic Regression is recommended for churn prediction, while
K-Means Clustering is recommended for customer segmentation and
targeted marketing strategies.
